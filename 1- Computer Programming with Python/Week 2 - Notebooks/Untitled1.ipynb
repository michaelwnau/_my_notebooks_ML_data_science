```python
import pandas as pd
```

```python
con = sqlite3.connect('sakila.db')
```

```python
def sql_to_df(sql_query):
    df = pd.read_sql(sql_query, con)
    return df
```

```python
query = '''
    SELECT *
    FROM sqlite_master
    WHERE type = 'table';
'''

tables = sql_to_df(query)
tables
```

```python
query = '''
    SELECT first_name, last_name
    FROM customer
'''

customer_names = sql_to_df(query)
customer_names
```

```python
query = '''
    SELECT *
    FROM film
    WHERE description
    LIKE '%Pastry%'
'''

pastry_films = sql_to_df(query)
pastry_films
```

```python
query = '''
    SELECT
        COUNT(title) AS Count,
        rating
    FROM film
    WHERE description
    LIKE '%Pastry%'
    GROUP BY rating
    ORDER BY Count DESC;
'''
pastry_films_by_rating = sql_to_df(query)
pastry_films_by_rating
```
